```python
"""
الف( تصویر را از این لینک دریافت کنید و با استفاده از انواع فیلترهای لبهیاب Roberts ،Prewitt ،Sobel، عمودی، Sobel
افقی، الپالسین و LOG لبهیابی را انجام دهید.
ب( لبه های این تصویر را با استفاده از الگوریتم لبه یابی کنی )Canny )و استفاده از مقادیر مناسب برای threshold ها لبه های
عکس ها را استخراج کنید و نتایج را نمایش دهید.
"""
```

```python
import numpy as np
import cv2
import matplotlib.image as mpli
import matplotlib.pyplot as plt
from scipy.ndimage import convolve, laplace

image0= mpli.imread('D:\\sadeneh\\تمرین دوم\\سادنه صادقی- 9913012008\\Q1\\1-1.jpg')
plt.subplot(2, 4, 8)
plt.imshow(image0, cmap='gray')
plt.title('Main')
plt.axis('off')
# خواندن عکس ورودی
image= image0.mean(axis=2) # تبدیل به فضای خاکستری

# Convert image to float64
image= image.astype(np.float64)


# Roberts Cross
# تعریف کرنل های عمودی و افقی
v = np.array([[1, 0], [0, -1]])# عمودی
h = np.array([[0, 1], [-1, 0]])# افقی
#اعمال کرنل های رابرتز در تصویر
vImg= convolve(image, v)
hImg= convolve(image, h)
imageRoberts = np.sqrt(vImg**2 + hImg**2)
#show image
plt.subplot(2, 4, 1)
plt.imshow(imageRoberts, cmap='gray')
plt.title('Roberts')
plt.axis('off')

# Sobel
# پیدا کردن لبه عمودی
v= cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
# پیدا کردن لبه افقی
h= cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
# ترکیب دو لبه عمودی و افقی
sobel= np.sqrt(v**2 + h**2)
#show image
plt.subplot(2, 4,2)
plt.imshow(v, cmap='gray')
plt.title('Sobel Vertical')
plt.axis('off')
#show image
plt.subplot(2, 4, 3)
plt.imshow(h, cmap='gray')
plt.title('Sobel Horizontal')
plt.axis('off')
#show image
plt.subplot(2, 4, 4)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel Combined')
plt.axis('off')



#Laplacian of Gaussian (LOG)
# اعمال فلیتر gaussian blur
blurred = cv2.GaussianBlur(image, (3, 3), 0)
#اعمال فیلتر گوسی بر تصویر مات شده
log= laplace(blurred)
#show image
plt.subplot(2, 4, 5)
plt.imshow(log, cmap='gray')
plt.title('Laplacian of Gaussian (LOG)')
plt.axis('off')


# Prewitt
#تعریف کرنل های پرویت
v = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitth_kernel_h = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
vImg = convolve(image, v)
hImg = convolve(image, h)
imagePrewitt = np.sqrt(vImg**2 + hImg**2)
#show image
plt.subplot(2, 4, 6)
plt.imshow(imagePrewitt, cmap='gray')
plt.title('Prewitt')
plt.axis('off')


# Laplacian
#اعمال فیلتر لاپلاسی در تصویر
laplacian = cv2.Laplacian(image, cv2.CV_64F)
#show image
plt.subplot(2, 4, 7)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')

plt.tight_layout()
plt.show()
```
