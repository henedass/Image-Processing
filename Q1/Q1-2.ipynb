```python
import matplotlib.image as mpli
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Function to apply Canny edge detection
def canny(image_path, threshold1, threshold2):
  
    image= mpli.imread(image_path)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    # If the image has an alpha channel, remove it
    if image.shape[-1] == 4:
        image= image[..., :3]
    
    # Convert the image to grayscale if it is in color
    if len(image.shape) == 3:
        image= image.mean(axis=2)
    
    # Normalize the image to the range [0, 255] and convert to uint8
    image= (image * 255).astype(np.uint8)
    
    # Apply Canny edge detection
    edges= cv2.Canny(image, threshold1, threshold2)
    
    # Plot the edge-detected image
   
    
    plt.subplot(1, 3, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edges ')
    plt.axis('off')
    
    plt.show()

# Path to the image
image_path= 'D:\\sadeneh\\تمرین دوم\\سادنه صادقی- 9913012008\\Q1\\1.2.jpg' 

# Apply the function with appropriate thresholds
canny(image_path, 50, 1000)  
canny(image_path, 50, 100) 
canny(image_path, 10, 2000) 
canny(image_path, 1000, 1000) 
```
