```python
"""
برای بهبود کنتراست دو روش کشش هیستوگرام و  
همسان سازی هیستوگرام وجود دارد که روش دوم برای همه حالت های گزینه بهتریست
باری این روش نیاز است هیستوگرام تصویر را بدست اورده و تابع تجمعی پیکسل ها را بدست اوریم
و همسان سازی را روی آنها اعمال کنیم

اعداد اعشاری با ازمایش و خطا بدست امده اند. با اعداد مختلف کنتراست تصویر و نمودار هیستوگرام ان تغییر میکند
"""

import numpy as np
import matplotlib.image as mpli
import matplotlib.pyplot as mplp


#Question 5

img1 = mpli.imread('D:\\sadeneh\\سادنه صادقی-9913012008\\Tree\\tree.jpg')
# تبدیل تصویر به مقیاس خاکستری با روش luminosity
gray =np.dot(img1[... ,:3],[0.1, 0.1, 1])  
#محاسبه نمودار هیستوگرام تصویر ورودی
histo1,bins=np.histogram(gray.flatten(), 256,[0,256])
#محاسبه تابع توزیع تجمعی
cdf=histo1.cumsum() 
#نرمال کردن تابع تجمعی- بازه اعداد بین صفر و بیشترین مقدار هیستوگرام است.
norm =(cdf*histo1.max()) / cdf.max() 
#همسان سازی هیستوگرام با کمک درونیابی خطی
img2 =np.interp(gray.flatten(), bins[:-1], norm).reshape(gray.shape)
#هیستوگرام تصویر بهبود یافته
histo2, _ =np.histogram(img2.flatten(), 256, [0,256])
#نمایش تصویر خروجی
mplp.imshow(img2, cmap ='gray')
mplp.axis('off')
mplp.show()


#نمایش هیستوگرام تصویر اولیه
#mplp.plot(inputHisto, color ='blue')


#نمایش هیستوگرام تصویر خروجی
mplp.plot(histo2)
mplp.show()
```
