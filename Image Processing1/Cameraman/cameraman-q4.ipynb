```python
"""
برای دوبرابر کردن تصویر سه روش وجود داشت. ساد ترین ان درونیابی نزدیکترین همسایه است که کیفیت پایین ری دارد. 
دو روش دیگر درونیابی دوخطی و دومکعبی بود که پیچیدگی محساباتی بالا ولی کیفیت بهتر داشتند
"""
import numpy as np
import matplotlib.pyplot as mplp
import matplotlib.image as mpli
#ًQuastion 4


# خواندن تصویر
img0 = mpli.imread( 'D:\\sadeneh\\سادنه صادقی-9913012008\\Cameraman\\cameraman.jpeg')

print( img0.shape)


# دریافت ابعاد تصویر
h0, w0,c0= img0.shape #h= height   w=width   c=channel

# دو برابر کردن اندازه
w1 = w0 * 2
h1 = h0 * 2


# ایجاد آرایه جدید برای تصویر تغییر اندازه داده شده
img1 = np.zeros((h1, w1, c0), dtype=img0.dtype) #img1= output image

# پر کردن آرایه با استفاده از نزدیک‌ترین همسایه
for x in range( h1):
    for y in range(w1 ):
        img1[x, y ] = img0[ min(x // 2, h0 - 1), min(y// 2, w0 - 1)]


mplp.imshow(img1 )
mplp.axis('off')
mplp.show()
print( img1.shape)
```
