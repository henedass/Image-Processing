```python
'''
برای چرخش 90 درجه و 180درجه تنها از معکوس کردن سطر و ستون ها استفاده میکنیم.
در چرخش 90درجه ستون ها رو معکوس میکنیم ولی سطرها را تغییر نمیدهیم.

در چرخش 180  علاوه بر معکوس کردن ستون ها باید سطرها را معکوس کنیم تا تصویر 90 درجه دیگر بچرخد

برای چرخش 45 درجه باید یک تابع  مجزا بنویسیم
در اینجا هر پیکسل را به مرکز تصویر منتقل مکینیم
و با کمک روابط چرخش مختصات جدیدش را حساب میکنیم
اگر مختصات جدید در محدوده تصویر ابعاد جدید باشد پیکسل مروبطه را در تصویر قرار میدیم

'''
import numpy as np
import matplotlib.pyplot as mplp
import matplotlib.image as mpli


#Question1


# Read the image path
inputImage= mpli.imread('D:\\sadeneh\\سادنه صادقی-9913012008\\Cameraman\\cameraman.jpeg')

#Fuction for 45  degree Rotaion
def degree45( inputImage ):
    h,w,c=inputImage.shape #get the dimension of image   #h=height    w= width  c=channel
    #create new demnsion for the  rotated image with the help of square diameter
    newDim=int(np.ceil(np.sqrt(h**2 +w**2))) 
    img1=np.zeros((newDim , newDim , c ) ,dtype = inputImage.dtype ) #create a blank image for the rotated image
    
    orgX , orgY = w//2 , h//2 #تعیین مختصات مرکزی تصویر ورودی
    orgX1 , orgY1 = newDim//2 , newDim//2 # تعیین مختصات مرکزی تصویر خروجی
    

    
    #convert degree to radians
    angle=np.deg2rad( 45 )
    #claculate cosin and sin of thr rotation angle
    cosAngle=np.cos( angle )
    sinAngle=np.sin( angle )

    #a loop for rotating the original pixels and place it in the new image
    for i in range ( h ):
        for j in range( w ):
            x = j-orgX
            y = i-orgY
            newX = int( cosAngle * x - sinAngle*y) + orgX1
            newY = int( sinAngle * x + cosAngle*y ) + orgY1
            if 0<=newX <newDim and 0<=newY<newDim:
                img1[ newY,newX ]= inputImage[ i,j ] 
    return img1
#apply rotation to the image
rotated45Image=degree45( inputImage ) 
mplp.axis('off')
mplp.imshow( rotated45Image )
```
