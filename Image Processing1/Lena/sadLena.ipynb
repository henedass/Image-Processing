```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

# بارگذاری تصویر با PIL
path = 'D:\\sadeneh\\سادنه صادقی-9913012008\\Lena\\lenna.jpg'
img1 = Image.open(path).convert('L')  # تبدیل تصویر به حالت خاکستری

# تبدیل تصویر به آرایه numpy
img2= np.array(img1)

# بررسی ابعاد تصویر
width ,height = img2.shape

# ایجاد نقشه تغییر شکل
map_x ,map_y= np.indices((width,height), dtype=np.float32)
center_x,center_y=  width // 2,height // 2

# پارامترهای انحنا
curve = 20
  # افزایش مقدار انحنا

# محاسبه انحنا در وسط تصویر (برای انحنای به سمت پایین)
map_x += curve * np.sin(2 * np.pi * (map_y - center_y) / width)

# اعمال تغییر شکل به تصویر
img3 = cv2.remap(img2, map_y, map_x, interpolation=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT)

# نمایش تصویر اصلی و تصویر تغییر شکل یافته
plt.imshow(img3, cmap='gray')
plt.show()
```
